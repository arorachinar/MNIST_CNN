```python
import tensorflow as tf
```

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist=input_data.read_data_sets("MNIST_data",one_hot=True)
```

### What we will be doing
1. Write a few helper functions(init_wgts,init_bias,conv2d)
2. variables,placeholders,cost,optimizers
3. Train CNN

```python
def init_wgts(shape):
    init_random_dist=tf.truncated_normal(shape,stddev=0.1)
    return tf.Variable(init_random_dist)
```

```python
def init_bias(shape):
    init_bias_vals=tf.constant(0.1,shape=shape)
    return tf.Variable(init_bias_vals)
```

```python
def conv2d(x,W):
    """
    x->[batch,H,W,Channels]
    W->[filter H,Filter W, Channel IN,Channel OUT]
    Use a stride of 1,keep padding same
    """
    return tf.nn.conv2d(x,W,strides=[1,1,1,1],padding='SAME')
```

```python
def max_pool_2by2(x):
    """
    x->[batch,h,w,channels], a 4d tensor
    """
    return tf.nn.max_pool(x,ksize=[1,2,2,1],strides=[1,2,2,1],padding='SAME')
```

```python
def convolutional_layer(input_x,shape):
    W=init_wgts(shape)
    b=init_bias([shape[3]])
    return tf.nn.relu(conv2d(input_x,W)+b)
```

```python
def normal_full_layer(input_layer,size):
    input_size=int(input_layer.get_shape()[1])
    W=init_wgts([input_size,size])
    b=init_bias([size])
    return tf.matmul(input_layer,W)+b
```

```python
#Placeholder
x=tf.placeholder(tf.float32,shape=[None,784])
y_true=tf.placeholder(tf.float32,shape=[None,10])
```

```python
x_image=tf.reshape(x,[-1,28,28,1])
```

```python
#Layer 1
convo_1=convolutional_layer(x_image,shape=[5,5,1,32])
convo_1_pooling=max_pool_2by2(convo_1)
#Layer 2
convo_2=convolutional_layer(convo_1_pooling,shape=[5,5,32,64])
convo_2_pooling=max_pool_2by2(convo_2)
#flatten it out
convo_2_flat=tf.reshape(convo_2_pooling,[-1,7*7*64])
full_layer_one=tf.nn.relu(normal_full_layer(convo_2_flat,1064))
```

```python
hold_prob=tf.placeholder(tf.float32)
full_one_dropout=tf.nn.dropout(full_layer_one,keep_prob=hold_prob)
```

```python
y_pred=normal_full_layer(full_one_dropout,10)
```

```python
cross_entropy=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(labels=y_true,logits=y_pred))
```

```python
#optimiser
optimizer=tf.train.AdamOptimizer(learning_rate=0.0001)
train=optimizer.minimize(cross_entropy)
```

```python
tf.reset_default_graph()
```

```python
init=tf.global_variables_initializer()
```

```python

#sess.run(train,feed_dict={x:batch_x,y:batch_y,hold_prob=0.45})

steps=5000
```

```python
session=tf.Session()
with session as sess:
    sess.run(init)
    for i in range(steps):
        batch_x,batch_y=mnist.train.next_batch(500)
        sess.run(train,feed_dict={x:batch_x,y_true:batch_y,hold_prob:0.5})
        if i%100==0:
            print("Currently on step{}".format(i))
            print("Accuracy is:")
            matches=tf.equal(tf.argmax(y_pred,1),tf.argmax(y_true,1))
            acc=tf.reduce_mean(tf.cast(matches,tf.float32))
            print(sess.run(acc,feed_dict={x:mnist.test.images,y_true:mnist.test.labels,hold_prob:1.0}))
            print('\n')
```
